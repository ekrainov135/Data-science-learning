```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
flights_data = sns.load_dataset("flights")
flights_data.head()
```

```python
sns.scatterplot(data=flights_data, x="year", y="passengers")
```

```python
sns.lineplot(data=flights_data, x="year", y="passengers");
```

```python
sns.relplot(data=flights_data, x="year", y="passengers", hue="month", kind="line");
```

```python
sns.relplot(data=flights_data, x="month", y="passengers", hue="year", kind="line");
```

```python
sns.barplot(data=flights_data, x="year", y="passengers")
plt.xticks(rotation = 45);
```

```python
uniform_data = np.random.rand(10, 12)
sns.heatmap(uniform_data);
```

```python
data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size=2000)
data = pd.DataFrame(data, columns=['x', 'y'])

for col in 'xy':
    plt.hist(data[col], alpha=0.5)
```

```python
for col in 'xy':
    sns.kdeplot(data[col], shade=True)
```
